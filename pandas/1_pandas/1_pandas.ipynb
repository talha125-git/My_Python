```python
#  filtering data (conditions)

mask = data["city"] == "Hyderabad"
mask
```

```python
# take data without true false
data[mask]
```

```python
# how many match played in Hyderabad
played = data[mask].shape[0]
print("Matches played in Hyderabad = ", played)
```

```python
# make function for cities 
def get_city_match_count(city):
    mask = data['city'] == city
    return data[mask].shape[0]
```

```python
# call function by taking city name 
print("matches played in =", get_city_match_count("Bangalore"))

# data["city"]
```

```python
get_city_match_count("Pune")
```

```python
#  matches play in hyderabad after 2010

mask1 = data["city"] == "Hyderabad"
mask2 = data["date"] > "2010-01-01"

data[mask1 & mask2].shape[0]
```

```python
def get_data_after_2010(city):
    mask1 = data["city"] == city
    mask2 = data["date"] > "2010-01-01"
    return data[mask1 & mask2].shape[0]

get_data_after_2010("Bangalore")
```

```python
get_data_after_2010("Pune")
```

```python
print("Total matches Played in Hyderabad      =", get_city_match_count("Hyderabad"))
print("Matches played after 2010 in Hyderabad =", get_data_after_2010("Hyderabad"))
```
