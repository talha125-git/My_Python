```python
# Against which teams has Virat Kohli scored the most runs
vk = delivery[delivery['batsman'] == "V Kohli"]
vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
# Make a function where we enter a player’s name and it shows the 
# bowling team against which that player has scored the most runs.
def run_scored(batsman_name):
    vk = delivery[delivery['batsman'] == batsman_name]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).index[0]

run_scored("V Kohli")
```

```python
print(run_scored("DA Warner"))
print(run_scored('G Gambhir'))
print(run_scored('MS Dhoni'))
```

```python
# show those who played over 200 balls
death_over = delivery[delivery['over'] > 15 ]
```

```python
all_batsman = death_over.groupby('batsman')['batsman_runs'].count()
x = all_batsman > 200
batsman_lsit = all_batsman[x].index.to_list()
```

```python
final = delivery[delivery['batsman'].isin(batsman_lsit)]
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
# strikr rate
sr = (runs/balls)*100
sr.sort_values(ascending=False)
```

```python
match = pd.read_csv('matches.csv')
```

```python
# merge
new =  delivery.merge(match, left_on='match_id', right_on='id')
print(new.shape)
print(match.shape)
```

```python
new.groupby(['season','batsman'])['batsman_runs'].sum()
# convert series to data frames use -> reset_index()
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index()
```

```python
#  remove duplication
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season', keep='first')
#  sort season col
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season', keep='first').sort_values('season')
```

```python
# sjow just season and batsman col
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season', keep='first').sort_values('season')[['season', 'batsman']]
```
